**Задачи (в сумме 5 баллов):**
- допишите метод `occypy`, чтобы он проверял занятость номера перед бронировнием: если номер занят, то генерируется исключение `RuntimeError`
- напишите метод `__str__` для красивой печати списка номеров через вызов функции `print`
- напишите метод, который занимает все *свободные* номера в отеле
- напишите метод, который возвращает долю занятых номеров
- напишите метод, который освобождает все занятые номера отеля
- напишите метод, который возвращает выручку, исходя из занятости комнат (стоимость комнаты - 5 тыс. у.е.)
- проверьте созданные методы через создание объектов и их вызов

```python
class Hotel:
    def __init__(self, num_rooms):
        self._rooms = [0 for _ in range(num_rooms)] # информация о статусе номеров

    # метод для бронирования номера по уникальному значению в списке
    def occypy(self, room_id):
      try:  
        if self._rooms[room_id - 1] == 1: # бронируем номер (вряд ли в отеле есть номера с нулем, поэтому дадим возможность менеджеру заполнять не с 0, а с 1)
          raise RuntimeError
        else:
          self._rooms[room_id - 1] = 1
      except RuntimeError:
        print("Ooops, номер уже занят!")

    # метод для освобождения номера по уникальному значению в списке
    def free(self, room_id):
        self._rooms[room_id - 1] = 0  # освобождаем номер

    # метод для красивого имени
    def __str__(self, num_rooms):
      return f"{self.num_rooms}"
    
    # метод для заполнения незанятых номеров
    def allbusy(self, _rooms):
      for i in range(len(_rooms)):
        if _rooms[i] != 1:
          _rooms[i] = 1
    
    # метод для вывода доли занятых номеров
    def part(self, _rooms):
      count = 0
      for i in range(len(_rooms)):
        if _rooms[i] == 1:
          count += 1
      print(f'{round(count/len(_rooms)*100)}% - доля занятых номеров')

    # метод для "очистки" занятых номеров
    def allfree(self, _rooms):
      for i in range(len(_rooms)):
        if _rooms[i] != 0:
          _rooms[i] = 0
    
    # метод для расчёта выручки
    def revenue(self, _rooms):
      sum = 0
      for i in range(len(_rooms)):
        if _rooms[i] != 0:
          sum += 5000
      print(f'Выручка = {sum} у. е.')
    
    
hotel = Hotel(5)
print(hotel._rooms)
hotel.occypy(3)
print(hotel._rooms)
hotel.part(hotel._rooms)
hotel.revenue(hotel._rooms)
hotel.free(3)
print(hotel._rooms)
hotel.allbusy(hotel._rooms)
print(hotel._rooms)
hotel.allfree(hotel._rooms)
print(hotel._rooms)
```

```python
class Hotel:
    def __init__(self, num_SGL = 0, num_DBL = 0, num_jun = 0, num_suite = 0):
        self.num_SGL = num_SGL
        self.num_DBL = num_DBL
        self.num_jun = num_jun
        self.num_suite = num_suite
        self._rooms = {"SGL": {"num" : num_SGL, "cost" : 5000, "busy" : [0 for _ in range(num_SGL)]},
                       "DBL" : {"num" : num_DBL, "cost" : 7000, "busy" : [0 for _ in range(num_DBL)]},
                       "Junior Suite" : {"num" : num_jun, "cost" : 4000, "busy" : [0 for _ in range(num_jun)]},
                       "Suite" : {"num" : num_suite, "cost" : 6500, "busy" : [0 for _ in range(num_suite)]}}

    def __str__(self, _rooms):
      return f"{self._rooms}"

    # метод для бронирования номера по уникальному значению в списке
    def occypy(self, typenum, room_id):
      try:  
        if self._rooms[typenum]["busy"][room_id - 1] == 1: # бронируем номер (вряд ли в отеле есть номера с нулем, поэтому дадим возможность менеджеру заполнять не с 0, а с 1)
          raise RuntimeError
        else:
          self._rooms[typenum]["busy"][room_id - 1] = 1
          return print(f"Вы забронировали 1 номер {typenum} за {self._rooms[typenum]['cost']} рублей")
      except RuntimeError:
        print("Ooops, номер уже занят!")
    
    # метод для освобождения номера по уникальному значению в списке
    def free(self, typenum, room_id):
      print("\n")
      try:  
        if self._rooms[typenum]["busy"][room_id - 1] != 1:
          raise RuntimeError
        else:
          self._rooms[typenum]["busy"][room_id - 1] = 0
          return print(f"Вы освободили 1 номер {typenum}")
      except RuntimeError:
        print("Ooops, номер уже свободен!")

    # метод для заполнения незанятых номеров опред типа
    def allbusytype(self, typenum):
      for i in range(len(self._rooms[typenum]["busy"])):
        if self._rooms[typenum]["busy"][i] != 1:
          self._rooms[typenum]["busy"][i] = 1
    
    # метод для заполнения незанятых номеров всего
    def allbusy(self):
      spis = ["SGL", "DBL", "Junior Suite", "Suite"]  # создаём список всех типов
      for j in range(len(spis)): # повторяем 4 раза - список из 4 элементов
        for i in range(len(self._rooms[spis[j]]["busy"])):
          if self._rooms[spis[j]]["busy"][i] != 1:
            self._rooms[spis[j]]["busy"][i] = 1
    
    # метод для "очистки" незанятых номеров опред типа
    def allfreetype(self, typenum):
      for i in range(len(self._rooms[typenum]["busy"])):
        if self._rooms[typenum]["busy"][i] == 1:
          self._rooms[typenum]["busy"][i] = 0

    # метод для "очистки" незанятых номеров всего
    def allfree(self):
      spis = ["SGL", "DBL", "Junior Suite", "Suite"]
      for j in range(len(spis)):
        for i in range(len(self._rooms[spis[j]]["busy"])):
          if self._rooms[spis[j]]["busy"][i] == 1:
            self._rooms[spis[j]]["busy"][i] = 0
    
    # метод для вывода доли занятых номеров в разрезе типа
    def part(self):
      print("\n")
      # в countlocal - кол-во определённого типа, а в countall - накопление кол-ва всех мест по всем типам, countbusy - кол-во занятых всех мест
      countlocal, countall, countbusy = 0, 0, 0 
      spis = ["SGL", "DBL", "Junior Suite", "Suite"] # создаём список всех типов
      for j in range(len(spis)): # повторяем 4 раза - список из 4 элементов
        for i in range(len(self._rooms[spis[j]]["busy"])):
          if self._rooms[spis[j]]["busy"][i] == 1:
            countlocal += 1
          countall += 1
        print(f'{round(countlocal/len(self._rooms[spis[j]]["busy"])*100)}% - доля занятых номеров {spis[j]}')
        countbusy += countlocal
        countlocal = 0
      print(f'{round(countbusy/countall*100)}% - доля всех занятых номеров')

    # метод для расчёта выручки
    def sum_all(self):
      print("\n")
      sumlocal, sumall = 0, 0 # в sumlocal - сумма определённого типа, а в sumall - накопление суммы по всем типам
      spis = ["SGL", "DBL", "Junior Suite", "Suite"] # создаём список всех типов
      for j in range(len(spis)): # повторяем 4 раза - список из 4 элементов
        for i in range(len(self._rooms[spis[j]]["busy"])):
          if self._rooms[spis[j]]["busy"][i] == 1:
            sumlocal +=  self._rooms[spis[j]]["cost"]
          sumall += self._rooms[spis[j]]["cost"]
        print(f'{sumlocal} - выручка от занятых номеров {spis[j]}')
        sumlocal = 0
      print(f'{sumlocal} - выручка от всех занятых номеров')

# проверка
hotel1=Hotel(2,3,4,5)
hotel1.allbusy()
print(hotel1._rooms)
hotel1.allfree()
print(hotel1._rooms)
hotel1.occypy("DBL", 1)
print(hotel1._rooms)
hotel1.allbusytype("Junior Suite")
print(hotel1._rooms)
hotel1.part()
hotel1.sum_all()
hotel1.free("DBL", 1)
hotel1.allfreetype("Junior Suite")
print(hotel1._rooms)

```
